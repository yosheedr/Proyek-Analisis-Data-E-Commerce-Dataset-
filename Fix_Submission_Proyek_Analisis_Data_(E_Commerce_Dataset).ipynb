### Pertanyaan 2: Seiring berjalannya waktu, bagaimana kinerja penjualan platform E-Commerce kami?

```python
dataFrameMonthly = semuaData.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
dataFrameMonthly.index = dataFrameMonthly.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
dataFrameMonthly = dataFrameMonthly.reset_index()
dataFrameMonthly.rename(columns={
    "order_id": "order_count",
}, inplace=True)
dataFrameMonthly.head()
```

```python
dataFrameMonthly = dataFrameMonthly.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
dataFrameMonthly.head()
```

```python
dataFrameMonthly.sort_values(by='order_count')
```

```python
monthMapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

dataFrameMonthly["month_numeric"] = dataFrameMonthly["order_approved_at"].map(monthMapping)
dataFrameMonthly = dataFrameMonthly.sort_values("month_numeric")
dataFrameMonthly = dataFrameMonthly.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    dataFrameMonthly["order_approved_at"],
    dataFrameMonthly["order_count"],
    marker='o',
    linewidth=2,
    color="#068DA9"
)
plt.title("Angka Orders per Bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

Bulan September menunjukkan penurunan yang signifikan, dan bulan November menunjukkan kenaikan yang signifikan.

### Pertanyaan 4 : Bagaimana kepuasan pelanggan dengan layanan kami?


```python
reviewScores = semuaData['review_score'].value_counts().sort_values(ascending=False)

most_common_score = reviewScores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(x=reviewScores.index,
            y=reviewScores.values,
            order=reviewScores.index,
            palette=["#068DA9" if score == most_common_score else "#D3D3D3" for score in reviewScores.index]
            )

plt.title("Penilaian oleh pelanggan atas layanan kami", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(fontsize=12)
plt.show()
```

Grafik di atas menunjukkan bahwa pelanggan sangat puas dengan layanan yang diberikan; data menunjukkan bahwa pelanggan dengan rating 5 memiliki jumlah data yang paling banyak.

### Pertanyaan 5 : Apakah ada perbedaan antara demografi pelanggan kami dan preferensi pembelian mereka?


### Berdasarkan customer_state

```python
dataFrameBystate = semuaData.groupby(by="customer_state").customer_id.nunique().reset_index()
dataFrameBystate.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
dataFrameBystate.head()
```

### Pertanyaan 6 : Di mana lokasi geografis memiliki jumlah pelanggan tertinggi?

```python
geoOtherState = dataFrameGeo.groupby(['geolocation_zip_code_prefix'])['geolocation_state'].nunique().reset_index(name='count')
geoOtherState[geoOtherState['count']>= 2].shape
max_state = dataFrameGeo.groupby(['geolocation_zip_code_prefix','geolocation_state']).size().reset_index(name='count').drop_duplicates(subset = 'geolocation_zip_code_prefix').drop('count',axis=1)
```

```python
geoSilver = dataFrameGeo.groupby(['geolocation_zip_code_prefix','geolocation_city','geolocation_state'])[['geolocation_lat','geolocation_lng']].median().reset_index()
geoSilver = geoSilver.merge(max_state,on=['geolocation_zip_code_prefix','geolocation_state'],how='inner')
```

```python
custSilver = dataFrameCustomers.merge(geoSilver,left_on='customer_zip_code_prefix',right_on='geolocation_zip_code_prefix',how='inner')
```

```python
custSilver.head()
```

```python
custSilver.to_csv("geoLoc.csv", index=False)
```

```python
def plot_brazil_map(data):
    brazil = mpimg.imread(urllib.request.urlopen('https://i.pinimg.com/originals/3a/0c/e1/3a0ce18b3c842748c255bc0aa445ad41.jpg'),'jpg')
    ax = data.plot(kind="scatter", x="geolocation_lng", y="geolocation_lat", figsize=(10,10), alpha=0.3,s=0.3,c='maroon')
    plt.axis('off')
    plt.imshow(brazil, extent=[-73.98283055, -33.8,-33.75116944,5.4])
    plt.show()
```

```python
plot_brazil_map(custSilver.drop_duplicates(subset='customer_unique_id'))
```
